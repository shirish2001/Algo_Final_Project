```python
import random
import math
import matplotlib.pyplot as plt
import time
import networkx as nx

class TSPSolver:
    def __init__(self, coord_x, coord_y, pop_size, iterations, alpha, beta, evaporation_rate, pheromone_deposit):
        self.coord_x = coord_x
        self.coord_y = coord_y
        self.pop_size = pop_size
        self.iterations = iterations
        self.alpha = alpha
        self.beta = beta
        self.evaporation_rate = evaporation_rate
        self.pheromone_deposit = pheromone_deposit
        self.cities = list(zip(coord_x, coord_y))
        self.num_cities = len(self.cities)
        self.distance_matrix = self.initialize_distance_matrix()
        self.pheromone_matrix = [[1 for _ in range(self.num_cities)] for _ in range(self.num_cities)]
        self.best_path = None
        self.best_length = float('inf')

    def initialize_distance_matrix(self):
        dis = [[0 for _ in range(self.num_cities)] for _ in range(self.num_cities)]
        for i in range(self.num_cities):
            for j in range(i, self.num_cities):
                dis[i][j] = dis[j][i] = math.hypot(self.cities[j][0] - self.cities[i][0], self.cities[j][1] - self.cities[i][1])
        return dis

    def roulette_selection(self, probabilities, total):
        threshold = random.uniform(0, total)
        for i, weight in enumerate(probabilities):
            threshold -= weight
            if threshold <= 0:
                return i

    def construct_path(self):
        path = [0]
        unvisited = set(range(1, self.num_cities))
        while unvisited:
            current_city = path[-1]
            probabilities = [self.pheromone_matrix[current_city][next_city] ** self.alpha *
                             (1 / self.distance_matrix[current_city][next_city]) ** self.beta
                             for next_city in unvisited]
            total = sum(probabilities)
            next_city_index = self.roulette_selection(probabilities, total)
            next_city = list(unvisited)[next_city_index]
            path.append(next_city)
            unvisited.remove(next_city)
        path.append(0)
        return path

    def apply_2_opt(self, path):
        path_length = self.calculate_path_length(path)
        for i in range(1, len(path) - 2):
            for j in range(i + 1, len(path) - 1):
                new_path = path[:i] + path[i:j+1][::-1] + path[j+1:]
                new_length = self.calculate_path_length(new_path)
                if new_length < path_length:
                    return self.apply_2_opt(new_path)
        return path


    def calculate_path_length(self, path):
        return sum(self.distance_matrix[path[i]][path[i+1]] for i in range(len(path) - 1))

    def update_pheromone(self, paths, lengths):
        for i in range(self.num_cities):
            for j in range(self.num_cities):
                self.pheromone_matrix[i][j] *= (1 - self.evaporation_rate)
        for path, length in zip(paths, lengths):
            for i in range(len(path) - 1):
                self.pheromone_matrix[path[i]][path[i+1]] += self.pheromone_deposit / length

    def solve_ant_colony_optimizer(self):
        start_time = time.time()

        for _ in range(self.iterations):
            all_paths = [self.construct_path() for _ in range(self.pop_size)]
            optimized_paths = [self.apply_2_opt(path) for path in all_paths]
            optimized_lengths = [self.calculate_path_length(path) for path in optimized_paths]
            self.update_pheromone(optimized_paths, optimized_lengths)

            min_length = min(optimized_lengths)
            if min_length < self.best_length:
                self.best_length = min_length
                self.best_path = optimized_paths[optimized_lengths.index(min_length)]

        end_time = time.time()
        self.execution_time = end_time - start_time
        return self.best_path, self.best_length, self.execution_time

    def create_complete_graph(self, num_nodes, distances):
        G = nx.Graph()
        for i in range(num_nodes):
            for j in range(i + 1, num_nodes):
                G.add_edge(i, j, weight=distances[i][j])
        return G

    def solve_christofides_algorithm(self):
        start_time = time.time()

        G = self.create_complete_graph(self.num_cities, self.distance_matrix)
        T = nx.minimum_spanning_tree(G)
        odd_degree_nodes = [node for node in T.nodes if T.degree(node) % 2 == 1]
        matching = nx.max_weight_matching(G.subgraph(odd_degree_nodes), maxcardinality=True)
        M = nx.MultiGraph(T)
        M.add_edges_from(matching)
        euler_tour = list(nx.eulerian_circuit(M))
        visited = set()
        hamiltonian_tour = [euler_tour[0][0]]
        for edge in euler_tour:
            if edge[1] not in visited:
                hamiltonian_tour.append(edge[1])
                visited.add(edge[1])
        total_distance = self.calculate_path_length(hamiltonian_tour)

        end_time = time.time()
        self.execution_time = end_time - start_time
        return hamiltonian_tour, total_distance, self.execution_time


    def plot_results(self, best_path_ant_colony=None, best_path_christofides=None):
        plt.figure(figsize=(10, 8))

        plt.scatter(self.coord_x, self.coord_y, color='black', s=100, zorder=5)

        if best_path_ant_colony is not None:
            for i in range(-1, len(best_path_ant_colony) - 1):
                start = best_path_ant_colony[i]
                end = best_path_ant_colony[i+1]
                plt.plot([self.coord_x[start], self.coord_x[end]],
                         [self.coord_y[start], self.coord_y[end]], color='blue', zorder=2)
                distance = self.distance_matrix[start][end]
                mid_x = (self.coord_x[start] + self.coord_x[end]) / 2
                mid_y = (self.coord_y[start] + self.coord_y[end]) / 2
                plt.text(mid_x, mid_y, f"{distance:.2f} ", color="purple", fontsize=8, zorder=5)
                plt.title('ACO Best_Distances')

        if best_path_christofides is not None:
            for i in range(-1, len(best_path_christofides) - 1):
                start = best_path_christofides[i]
                end = best_path_christofides[i+1]
                plt.plot([self.coord_x[start], self.coord_x[end]],
                         [self.coord_y[start], self.coord_y[end]], color='green', zorder=2)
                distance = self.distance_matrix[start][end]
                mid_x = (self.coord_x[start] + self.coord_x[end]) / 2
                mid_y = (self.coord_y[start] + self.coord_y[end]) / 2
                plt.text(mid_x, mid_y, f"{distance:.2f} ", color="purple", fontsize=8, zorder=5)
                plt.title('Christofides Best_Distances')


        offset = plt.gcf().get_size_inches()*plt.gca().get_window_extent().height/plt.gcf().dpi/len(self.coord_x)*3


        for i, (x, y) in enumerate(zip(self.coord_x, self.coord_y)):
            plt.text(x, y, str(i), color="white", fontsize=9, ha='center', va='center', zorder=10)

        plt.axis('equal')
        plt.show()


if __name__ == '__main__':
    random.seed(42)
    city_num = 30
    coord_x = [random.uniform(0, 100) for _ in range(city_num)]
    coord_y = [random.uniform(0, 100) for _ in range(city_num)]
    solver = TSPSolver(coord_x, coord_y, pop_size=10, iterations=100, alpha=1, beta=2, evaporation_rate=0.5, pheromone_deposit=100)
    best_path, best_length, execution_time = solver.solve_ant_colony_optimizer()
    print(f"Best path: {best_path}\nBest length: {best_length}\nExecution time: {execution_time}s")
    solver.plot_results(best_path_ant_colony=best_path)
    c_best_path, c_best_cost, c_execution_time = solver.solve_christofides_algorithm()
    print(f"Best path: {c_best_path}\nBest length: {c_best_cost}\nExecution time: {c_execution_time}s")
    solver.plot_results(best_path_christofides=c_best_path)
```
